## Ejercicio 04

Utilice el arreglo creado en el ejercicio anterior. Reconvierta el arreglo a uno de dimensión 5 columnas y 30 filas.

A partir de ese arreglo cree otro nuevo en el cual cada uno de los elementos de las filas es mayor que el promedio de la respectiva columna.

Muestre el arreglo obtenido y sus características.

```python
import numpy as np

def reconstruir_arreglo(arreglo_entrada):
    """Reconstruye un arreglo de 5 columnas y 30 filas a partir de otro arreglo."""
    # Reconstruir el arreglo a 5 columnas y 30 filas
    arreglo_reconstruido = arreglo_entrada.reshape(30, 5)

    return arreglo_reconstruido

def crear_arreglo_filtrado(arreglo):
    """Crea un nuevo arreglo donde cada elemento de las filas es mayor que el promedio de su columna."""
    promedio_columnas = np.mean(arreglo, axis=0)  # Calcular el promedio de cada columna
    arreglo_filtrado = np.where(arreglo > promedio_columnas, arreglo, promedio_columnas)

    return arreglo_filtrado

# Reconstruir el arreglo de 5 columnas y 30 filas a partir del arreglo_03
arreglo_reconstruido = reconstruir_arreglo(arreglo_03)

# Crear un nuevo arreglo filtrado donde cada elemento de las filas es mayor que el promedio de su columna
arreglo_filtrado = crear_arreglo_filtrado(arreglo_reconstruido)

# Mostrar el arreglo filtrado y sus características
print("Arreglo filtrado:")
print(arreglo_filtrado)
print("\nDimensiones del arreglo:", arreglo_filtrado.shape)
print("Tipo de datos del arreglo:", arreglo_filtrado.dtype)
print("Valor máximo en el arreglo:", np.max(arreglo_filtrado))
print("Valor mínimo en el arreglo:", np.min(arreglo_filtrado))
print("Promedio de todos los elementos del arreglo:", np.mean(arreglo_filtrado))
```

## Ejercicio 05

Utilice el arreglo creado en el ejercicio 03.

Ordene cada columna de forma descendente, de mayor a menor, de forma independiente. Luego cree un nuevo arreglo con la concatenación de todas las columnas ordenadas.

Muestre el resultado y las características del arreglo.

```python
import numpy as np

def ordenar_columnas_descendente(arreglo):
    """Ordena cada columna de un arreglo de forma descendente."""
    columnas_ordenadas = np.sort(arreglo, axis=0)[::-1]
    return columnas_ordenadas

def concatenar_columnas_ordenadas(arreglo):
    """Concatena todas las columnas ordenadas en un nuevo arreglo."""
    arreglo_concatenado = np.concatenate(arreglo, axis=0)
    return arreglo_concatenado

# Utilizar el arreglo creado en el ejercicio 03
arreglo_ordenado = ordenar_columnas_descendente(arreglo_03)

# Concatenar todas las columnas ordenadas en un nuevo arreglo
arreglo_concatenado = concatenar_columnas_ordenadas(arreglo_ordenado)

# Mostrar el resultado y sus características
print("Arreglo concatenado con columnas ordenadas:")
print(arreglo_concatenado)
print("\nDimensiones del arreglo:", arreglo_concatenado.shape)
print("Tipo de datos del arreglo:", arreglo_concatenado.dtype)
print("Valor máximo en el arreglo:", np.max(arreglo_concatenado))
print("Valor mínimo en el arreglo:", np.min(arreglo_concatenado))
print("Promedio de todos los elementos del arreglo:", np.mean(arreglo_concatenado))
```

```python
import numpy as np

def crear_arreglo_07():
    """Crea un nuevo arreglo de 10x100 a partir de la columna del arreglo_00 con números entre 0 y 360."""
    arreglo_00 = crear_arreglo()

    # Extraer la columna del arreglo_00 con números entre 0 y 360
    columna_grados = arreglo_00[:, 10]

    # Repetir la columna de grados 10 veces (una vez por cada fila) y 10 veces más (100 en total)
    nuevo_arreglo = np.tile(columna_grados, (10, 10))

    return nuevo_arreglo

def convertir_a_radianes(arreglo):
    """Convierte los valores de un arreglo de grados sexagesimales a radianes."""
    arreglo_radianes = np.radians(arreglo)
    return arreglo_radianes

# Crear un nuevo arreglo de 10x100 a partir de la columna del arreglo_00 con números entre 0 y 360
nuevo_arreglo_07 = crear_arreglo_07()

# Convertir los valores del nuevo arreglo a radianes
arreglo_radianes = convertir_a_radianes(nuevo_arreglo_07)

# Mostrar el resultado y sus características
print("Arreglo de grados a radianes:")
print(arreglo_radianes)
print("\nDimensiones del arreglo:", arreglo_radianes.shape)
print("Tipo de datos del arreglo:", arreglo_radianes.dtype)
print("Valor máximo en el arreglo:", np.max(arreglo_radianes))
print("Valor mínimo en el arreglo:", np.min(arreglo_radianes))
print("Promedio de todos los elementos del arreglo:", np.mean(arreglo_radianes))
```
